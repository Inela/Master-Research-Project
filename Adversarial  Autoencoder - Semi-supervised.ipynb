### My data Adversarial Autoencoder


```python
import tensorflow as tf
import numpy as np
import datetime
import os
import argparse

import pickle


pix_flatten = 784
n1 = 100
n2 = 100
latent_size = 10
batch_n =100
epochs = 10
learn_rate = 0.001
decay_rate_beta_1 = 0.9
n_labels = 5
labeled_num = 2500

x_in = tf.placeholder(dtype=tf.float32, shape=[batch_n, pix_flatten])
x_in_labeled = tf.placeholder(dtype=tf.float32, shape=[batch_n, pix_flatten])
y_input = tf.placeholder(dtype=tf.float32, shape=[batch_n, n_labels])
x_out = tf.placeholder(dtype=tf.float32, shape=[batch_n, pix_flatten])
prior_dist = tf.placeholder(dtype=tf.float32, shape=[batch_n, latent_size])
dist_label = tf.placeholder(dtype=tf.float32, shape=[batch_n, n_labels])


def nn_layer(x, dim_1, dim_2 , name):
 
    with tf.variable_scope(name, reuse=None):
        weights = tf.get_variable("weights", shape=[dim_1, dim_2],
                                  initializer=tf.random_normal_initializer(mean=0., stddev=0.01))
        bias = tf.get_variable("bias", shape=[dim_2], initializer=tf.constant_initializer(0.0))
        out = tf.add(tf.matmul(x, weights), bias, name='matmul')
        return out

def encoder(x, reuse=False, supervised=False):

    if reuse:
        tf.get_variable_scope().reuse_variables()
    with tf.name_scope('Encoder'):
        encoder_layer_1 = tf.nn.relu(nn_layer(x, pix_flatten, n1, 'encoder_layer_1'))
        encoder_layer_2 = tf.nn.relu(nn_layer(encoder_layer_1, n1, n2, 'encoder_layer_2'))
        latent_var = nn_layer(encoder_layer_2, n2, latent_size, 'encoder_latent_var')
    
        label_en = nn_layer(encoder_layer_2, n2, n_labels, 'encoder_label')
        if not supervised:
            soft_lab = tf.nn.softmax(logits=label_en, name='encoder_softmax_label')
        else:
            soft_lab = label_en
        return soft_lab, latent_var

    
def decoder(x, reuse=False):

    if reuse:
        tf.get_variable_scope().reuse_variables()
    with tf.name_scope('decoder'):
        decoder_layer_1 = tf.nn.relu(nn_layer(x, latent_size + n_labels, n2, 'decoder_layer_1'))
        decoder_layer_2 = tf.nn.relu(nn_layer(decoder_layer_1, n2, n1, 'decoder_layer_2'))
        decoder_out = tf.nn.sigmoid(nn_layer(decoder_layer_2, n1, pix_flatten, 'decoder_output'))
        return decoder_out

def dsc_label(x, reuse=False):
 
    if reuse:
        tf.get_variable_scope().reuse_variables()
    with tf.name_scope('discriminator_label'):
        dsc_lab_1 = tf.nn.relu(nn_layer(x, n_labels, n1, name='discriminator_label_layer_1'))
        dsc_lab_2 = tf.nn.relu(nn_layer(dsc_lab_1, n1, n2, name='discriminator_label_layer_2'))
        disc_lab_out = nn_layer(dsc_lab_2, n2, 1, name='discriminator_label_out')
        return disc_lab_out 

    

def dsc_latent(x, reuse=False):

    if reuse:
        tf.get_variable_scope().reuse_variables()
    with tf.name_scope('discriminator_normal'):
        dsc_normal_1 = tf.nn.relu(nn_layer(x, latent_size, n1, name='discriminator_normal_layer_1'))
        dsc_normal_2 = tf.nn.relu(nn_layer(dsc_normal_1, n1, n2,name='discriminator_normal_layer_2'))
        disc_latent_out = nn_layer(dsc_normal_2, n2, 1, name='discriminator_normal_out')
        return disc_latent_out 

 
    
def batch_run(x, y, batch_n):

    first_index = np.arange(labeled_num)
    index = np.random.permutation(first_index)[:batch_n]
    return x[index], y[index]

def batch_val(x, y, batch_n , batch):
    if batch == 0 :
        first_index = 0 
    else :
        first_index = batch* batch_n 
    second_index = first_index + batch_n 
    return x[first_index:second_index], y[first_index:second_index]

def error_fun(predicted , real):

    error = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=predicted, logits=real))
    return error


def model_train(train_model=True):

 
    with tf.variable_scope(tf.get_variable_scope()):
        encoder_output_label, encoder_output_latent = encoder(x_in)
        decoder_input = tf.concat([encoder_output_label, encoder_output_latent], 1)
        decoder_output = decoder(decoder_input)


    with tf.variable_scope(tf.get_variable_scope()):
        dsc_normal_prior = dsc_latent(prior_dist)
        dsc_normal_gen_enc = dsc_latent(encoder_output_latent, reuse=True)

    with tf.variable_scope(tf.get_variable_scope()):
        dsc_lab_prior = dsc_label(dist_label)
        dsc_lab_gen_enc = dsc_label(encoder_output_label, reuse=True)


    with tf.variable_scope(tf.get_variable_scope()):
        encoder_output_label_, _ = encoder(x_in_labeled, reuse=True, supervised=True)
        output_label_val = tf.concat([x_in_labeled , encoder_output_label_, y_input],1)
        

    correct_pred = tf.equal(tf.argmax(encoder_output_label_, 1), tf.argmax(y_input, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))

    autoencoder_error = tf.reduce_mean(tf.square(x_out - decoder_output))
    
    dsc_normal_error_prior =error_fun(tf.ones_like(dsc_normal_prior) ,dsc_normal_prior )
    dsc_normal_error_gen_enc = error_fun(tf.zeros_like(dsc_normal_gen_enc),dsc_normal_gen_enc )
    dsc_normal_error_total = dsc_normal_error_gen_enc + dsc_normal_error_prior

    dsc_lab_error_prior = error_fun(tf.ones_like(dsc_lab_prior) ,dsc_lab_prior )
    dsc_lab_error_gen_enc = error_fun(tf.zeros_like(dsc_lab_gen_enc), dsc_lab_gen_enc)
    dsc_lab_error_total =  dsc_lab_error_prior +  dsc_lab_error_gen_enc

    generator_latent_error =error_fun(tf.ones_like(dsc_normal_gen_enc), dsc_normal_gen_enc)
    generator_label_error = error_fun(tf.ones_like(dsc_lab_gen_enc), dsc_lab_gen_enc)
    generator_error_total =  generator_latent_error +  generator_label_error

    supervised_encoder_error = error_fun(y_input, encoder_output_label_)

    all_var = tf.trainable_variables()
    encoder_var = [var for var in all_var if 'encoder_' in var.name]
    dsc_latent_var = [var for var in all_var if 'discriminator_normal_' in var.name]
    dsc_label_var = [var for var in all_var if 'discriminator_label_' in var.name]
 

    # AdamOptimizer 
    adamopt_autoencode  = tf.train.AdamOptimizer(learning_rate=learn_rate,beta1=decay_rate_beta_1)
    autoencoder_optm =    adamopt_autoencode.minimize(autoencoder_error)
    adamopt_dsc_latent = tf.train.AdamOptimizer(learning_rate=learn_rate, beta1=decay_rate_beta_1)
    dsc_latent_optm      = adamopt_dsc_latent.minimize(dsc_normal_error_total, var_list=dsc_latent_var)
    adamoptm_dsc_lab = tf.train.AdamOptimizer(learning_rate=learn_rate,beta1=decay_rate_beta_1)
    dsc_label_optm     =  adamoptm_dsc_lab.minimize(dsc_lab_error_total, var_list=dsc_label_var)
    adamoptm_generator = tf.train.AdamOptimizer(learning_rate=learn_rate, beta1=decay_rate_beta_1)
    generator_optm    = adamoptm_generator .minimize(generator_error_total, var_list=encoder_var)
    adamopt_supervised = tf.train.AdamOptimizer(learning_rate=learn_rate, beta1=decay_rate_beta_1)
    supervised_encoder_optm   =  adamopt_supervised.minimize(supervised_encoder_error, var_list=encoder_var)
 
    init = tf.global_variables_initializer()

    saver = tf.train.Saver()
    step = 0
    with tf.Session() as sess:
        if train_model:
            model_save = 'C:/Users/beqaj/Desktop/KERAS CODE MRP/results/semi_supervised_adv_autoencoder/saved_model'
            sess.run(init)
        
            x_l, y_l = x_labeled , y_labeled 
            output_label_val_total = []
            for num_epoch in range(epochs):
                
                n_batches = int(len(x_u_labeled)/batch_n)
                
                print("--Epoch {}/{}-----".format(num_epoch, epochs))
                
                for train_batch in range(1 , n_batches):
                    random_prior_dist = np.random.randn(batch_n, latent_size) * 5.
                    prior_lab_dist = np.random.randint(low=0, high=5, size=batch_n)
                    prior_lab_dist = np.eye(n_labels)[prior_lab_dist]
                    
                    x_ul, _ = batch_val(x_u_labeled, y_u_labeled, batch_n , train_batch)
                    
                    x_lab, y_lab = batch_run(x_l, y_l, batch_n)
                    
                    sess.run(autoencoder_optm, feed_dict={x_in: x_ul, x_out: x_ul})
                    sess.run(dsc_latent_optm, feed_dict={x_in: x_ul, x_out: x_ul, prior_dist: random_prior_dist})
                    sess.run(dsc_label_optm,feed_dict={x_in: x_ul, x_out: x_ul, dist_label: prior_lab_dist})
                    sess.run(generator_optm, feed_dict={x_in: x_ul, x_out: x_ul})
                    sess.run(supervised_encoder_optm, feed_dict={x_in_labeled: x_lab, y_input: y_lab})
                    if train_batch % 15 == 0:
                        error_autoencoder,  generator_error, supervised_enc_error = sess.run(
                            [autoencoder_error, generator_error_total, supervised_encoder_error],
                            feed_dict={x_in: x_ul, x_out: x_ul,
                                       prior_dist: random_prior_dist, y_input: y_lab, x_in_labeled: x_lab,
                                       dist_label: prior_lab_dist})
                
                        print("Epoch running: {}, Batch: {}".format(num_epoch, train_batch))
                    
                        print("Autoencoder Error: {}".format(error_autoencoder))
                        print("Generator Error: {}".format(generator_error))
                        print("Supervised Encoder Error: {}".format(supervised_enc_error))
                        
                        
                    step += 1
                encoder_accuracy = 0
                num_batches = int(len(x_validation)/batch_n)
                for test_batch in range(num_batches):
             
                    x_lab, y_lab = batch_val(x_validation, y_validation, batch_n,test_batch)
                    batch_accuracy = sess.run(accuracy, feed_dict={x_in_labeled: x_lab, y_input: y_lab})
                    output_label_save = sess.run( output_label_val , feed_dict={x_in_labeled: x_lab, y_input: y_lab})
                    output_label_val_total = np.append( output_label_val_total ,  output_label_save)
                    
                    encod_label = sess.run(encoder_output_label_ , feed_dict={x_in_labeled: x_lab}) 
                
                    encoder_accuracy += batch_accuracy
                    #if test_batch % 15 == 0:
                       #print("Batch Running: {} ,  Batch Accuracy : {}".format(test_batch ,batch_accuracy))
                    
                encoder_accuracy /= num_batches
                output_val = open('output_val_{},{}'.format(num_epoch , test_batch), 'wb')
                pickle.dump(output_label_val_total, output_val)
                output_val.close()
                
                print("Encoder Classification Accuracy: {}".format(encoder_accuracy))
           
                saver.save(sess, save_path=model_save, global_step=step)
        else:
           
            saver.restore(sess, save_path=tf.train.latest_checkpoint(model_save))



if __name__ == '__main__':
    model_train()
```
